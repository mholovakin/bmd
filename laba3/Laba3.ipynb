<h1>Візуалізація даних за допомогою бібліотеки Seaborn<br/>Головакін М.А. БС-82</h1>

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
sns.distplot(tips['passengers'])
```

```python
sns.jointplot(x='year', y='passengers', data=tips)
```

```python
sns.pairplot(tips)
```

```python
sns.barplot(x='year', y='passengers', data=tips)
```

```python
import numpy as np
```

```python
sns.barplot(x='year', y='passengers', data=tips, estimator=np.std)
```

```python
sns.boxplot(x = 'year', y='passengers', data=tips, palette='rainbow')
```

```python
sns.stripplot(x='year', y='passengers', data=tips)
```

```python
sns.heatmap(tips.corr())
```

```python
g = sns.FacetGrid(tips.head(10), col='passengers', row='year')
```

```python
import matplotlib.pyplot as plt
g = sns.FacetGrid(tips.head(), col='year', row='passengers')
g = g.map(plt.hist, 'year')
```
